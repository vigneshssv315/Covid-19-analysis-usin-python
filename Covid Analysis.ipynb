```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], axis=1, inplace=True, errors='ignore')
```

```python
covid_df.head()
```

```python
covid_df['Date']= pd.to_datetime(covid_df['Date'], format= '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
#Active Cases
covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

```python
statewise  = pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise['Recovery Rate'] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise['Mortality Rate'] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#top 10 states with active cases
top_10_ac=covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
fig = plt.Figure(figsize=(16,9))
plt.title("Top_10_States_with_high_active_cases",size=25)
ax = sns.barplot(data=top_10_ac.iloc[:10],y="Active_cases",x="State/UnionTerritory")
plt.xlabel("States")
plt.ylabel("Active Cases")
#plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# top 10 states with high deaths
top_10_deaths=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig = plt.Figure(figsize=(18,5))
plt.title("Top_10_States_with_high_deaths",size=25)
ax = sns.barplot(data=top_10_deaths.iloc[:12],y="Deaths",x="State/UnionTerritory",linewidth=2,edgecolor = 'black')
plt.xlabel("States")
plt.ylabel("Deaths")
#plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Growth trend
fig = plt.figure(figsize=(12,6))
ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerela','Uttar Pradesh','Tamil Nadu'])],x='Date',y='Active_cases',hue='State/UnionTerritory')
ax.set_title("Top 5 effected states in India",size=16)
```

```python
#Remove rows where state = India
vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated' : 'Total'},inplace=True)
vaccine.head()
```

```python
#Most Vaccinated States
max_vac = vaccine.groupby("State")["Total"].sum().to_frame('Total')
max_vac = max_vac.sort_values("Total",ascending=False)[:5]
max_vac
```

```python
fig = plt.Figure(figsize=(10,5))
plt.title("Top 5 Vaccinated States in India",size=20)
x = sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='Black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
